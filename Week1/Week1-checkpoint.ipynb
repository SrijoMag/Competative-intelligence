```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10,5)
df=pd.read_excel("Competatives_Sales.xlsx")
df.head()
```

```python
print(df.shape)
df.info()
df.columns
```

```python
df.replace("-", np.nan, inplace=True)
df.replace("", np.nan, inplace=True)
```

```python
df.rename(columns={"Avalability": "Availability"}, inplace=True)
```

```python

df["Avalaibilty"] = (
    df["Avalaibilty"].astype(str).str.lower()
    .replace({
        "yes":"yes",
        "yea":"yes",
        "few left":"few left"
    })
)
```

```python
df["Delivery_Fee"] = (
    df["Delivery_Fee"].astype(str).str.lower()
)
```

```python
df["Current_Price"] = (
    df["Current_Price"].astype(str).str.replace("₹","",regex=False).str.replace(",","",regex=False)
)
df["Original_Price"] = (
    df["Original_Price"].astype(str).str.replace("₹","",regex=False).str.replace(",","",regex=False)
)
df["Current_Price"] = pd.to_numeric(df["Current_Price"],errors="coerce")
df["Original_Price"] = pd.to_numeric(df["Original_Price"],errors="coerce")
```

```python
df["Discont%"] = pd.to_numeric(df["Discont%"], errors="coerce")
df["Ratings"] = pd.to_numeric(df["Ratings"], errors="coerce")
```

```python
df["Reviews"] = (
    df["Reviews"].astype(str).str.replace(",","",regex=False)
)
df["Reviews"] = pd.to_numeric(df["Reviews"], errors="coerce")
```

```python
df.info()
df.describe()
```

```python
avg_discount_brand = (
    df.groupby("Brand Name")["Discont%"].mean().sort_values(ascending=False)
)
print(avg_discount_brand)
```

```python
avg_discount_brand.plot(kind="bar")
plt.title("Average Discount by Brand")
plt.ylabel("Discount %")
plt.show()
```

```python
avg_price_brand = (
    df.groupby("Brand Name")["Current_Price"].mean().sort_values(ascending=False)
)

print(avg_price_brand)
```

```python
avg_price_brand.plot(kind="bar")

plt.title("Average Price by Brand")
plt.ylabel("Price")
plt.show()
```

```python
platform_discount = (
    df.groupby("Platform")["Discont%"]
    .mean()
)

print(platform_discount)
```

```python
platform_price = (
    df.groupby("Platform")["Current_Price"].mean()
)
platform_price.plot(kind="bar")
plt.title("Average Price by Platform")
plt.ylabel("Price")
plt.show()
```

```python
brand_rating = (
    df.groupby("Brand Name")["Ratings"].mean().sort_values(ascending=False)
)
print(brand_rating)
```

```python
brand_rating.plot(kind="bar")
plt.title("Average Rating by Brand")
plt.ylabel("Rating")
plt.show()
```

```python
top_reviewed = (
    df.sort_values("Reviews",ascending=False)
      [["Brand Name","Reviews"]].head(10)
)
top_reviewed
```

```python
highest_discount = (
    df.sort_values("Discont%",ascending=False)
      [["Brand Name","Discont%"]].head(10)
)
highest_discount
```

```python
def segment(price):

    if price < 1500:
        return "Budget"
    elif price <= 5000:
        return "Mid Range"
    else:
        return "Premium"
df["Segment"] = df["Current_Price"].apply(segment)
df[["Brand Name","Current_Price","Segment"]].head()
```

```python
segment_count = df["Segment"].value_counts()
segment_count.plot(kind="pie",autopct="%1.1f%%")

plt.ylabel("")
plt.title("Product Segment Distribution")
plt.show()
```

```python
offer_analysis = df["Discount_Type"].value_counts()
print(offer_analysis)
```

```python
offer_analysis.plot(kind="bar")
plt.title("Offer Type Distribution")
plt.show()
```

```python
summary = df.groupby("Brand Name").agg({
    "Current_Price":"mean",
    "Discont%":"mean",
    "Ratings":"mean",
    "Reviews":"mean"
}).round(2)

summary
```

```python
summary.to_excel("Competitor_Summary.xlsx")
```
